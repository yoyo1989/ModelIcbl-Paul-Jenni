This notebook
- compare synthetic spectra (p02NiRcA_e.spec and p02NiRcA_p.spec) to average SN Ic-bl spectra without and with continuum
- compare sythetic LCs (p02NiRcA_e.lc and p02NiRcA_p.lc) to those of SN Ic-bl shown in figure 8 of [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf) 

Main conclusion
- Synthetic spectra from equatorial view angle and from polar view angle have a relatively large difference at phase -5 days.
- Synthetic spectra from ploar view angle can almost well reproduce average spectra (except for wavelengths larger than 7000 angstrom) till phase -15 days.
- Synthetic LCs are similar with LCs in figure 8 of [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf) 


```python
import numpy as np
from scipy.io.idl import readsav
import pylab as pl
import pandas as pd
import readcoll 
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from IPython.display import Image
from IPython.display import display
%matplotlib inline
```

# I. Plots below compare synthetic spectra to average SN Ic-bl spectra without and with continuum

## 1) Compare spectra without continuum

```python
def plotmont1(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        #print spec+phase[i]+'_flat.txt'
        w,f = readcoll.readcol(spec+phase[i]+'_flat.txt',twod=False)
        #print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        pl.text(4100,shift[i]+1.5,phasemean[i]+'+/-2d',fontsize=20)
        if i == 0:
            pl.plot(w, f*5+shift[i], label=spec, color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f*5+shift[i], color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3)    
        pl.fill_between(s.wlog, s.fmean*5-s.fsdev*5+shift[i], s.fmean*5+s.fsdev*5+shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont1('Dec2016/p02NiRcA_e', 
         ['9', '14', '19', '24', '29', '34', '39'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened (equatorial view angle) vs average IcBL',fontsize=20)

plt.subplot(122)
plotmont1('Dec2016/p02NiRcA_p', 
         ['8', '13', '18', '23', '28', '33', '38'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened (polar view angle) vs average IcBL',fontsize=20)
```

## 2) Compare spectra with continuum

```python
def plotmont2(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        #print spec+phase[i]+'.txt'
        w,f = readcoll.readcol(spec+phase[i]+'.txt',twod=False)
        f=f/np.mean(f)/4
        #print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        dwbin=s.wlog[1:1024]-s.wlog[0:1023] # very important
        dwbin=np.append(dwbin[0], dwbin)/10
        pl.text(5200,shift[i]-0.5,phasemean[i]+'+/-2d',fontsize=20)
        
        if i == 0:
            pl.plot(w, f+shift[i], label=spec, color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f+shift[i], color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3)   
        pl.fill_between(s.wlog, (s.fmean-s.fsdev+1)*s.smean/dwbin+shift[i], (s.fmean+s.fsdev+1)*s.smean/dwbin
                        +shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont2('Dec2016/p02NiRcA_e', 
         ['9', '14', '19', '24', '29', '34'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05','00', '05', '10', '15', '20'], 
         [6, 10, 14, 18, 22, 26, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 33)
pl.legend(fontsize=20)
plt.tick_params(labelsize=20)
pl.title('Original (equatorial view angle) vs average IcBL',fontsize=20)

plt.subplot(122)
plotmont2('Dec2016/p02NiRcA_p', 
         ['8', '13', '18', '23', '28', '33'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15', '20'], 
         [6, 10, 14, 18, 22, 26, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 33)
plt.tick_params(labelsize=20)
pl.title('Original (polar view angle) vs average IcBL',fontsize=20)
pl.legend(fontsize=20)
```

# II. Light curves

Synthetic LCs are very similar to LCs in figure 8 of [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf) 

```python
t,L = readcoll.readcol('Dec2016/p02NiRcA_e.lc',twod=False)
pl.plot(t/3600/24-13.5, np.log10(L), label='equatorial view angle')
t,L = readcoll.readcol('Dec2016/p02NiRcA_p.lc',twod=False)
pl.plot(t/3600/24-12.5, np.log10(L), label='polar view angle')
pl.xlabel("Days WRT bolometric max",fontsize=20)
pl.ylabel("Log(Luminosity)",fontsize=20)
#pl.xlim(-30, 100)
#pl.ylim(40.5, 43.5)
pl.legend(fontsize=15,loc='lower right')
pl.title('Synthetic Bolometric LC',fontsize=20)
```
