```python
import numpy as np
from scipy.io.idl import readsav
import pylab as pl
import pandas as pd
import readcoll 
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plotmont1(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        print spec+phase[i]+'_flat.txt'
        w,f = readcoll.readcol(spec+phase[i]+'_flat.txt',twod=False)
        print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        pl.text(4000,shift[i]+1,phasemean[i]+'+/-2d',fontsize=20)
        if i == 0:
            pl.plot(w, f*5+shift[i], label=spec, color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f*5+shift[i], color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3)    
        pl.fill_between(s.wlog, s.fmean*5-s.fsdev*5+shift[i], s.fmean*5+s.fsdev*5+shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont1('blicSpec_a_', 
         ['2', '6', '10', '16', '20', '26'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15', '20'], 
         [2, 6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(0, 31)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened blicSpec_a_ vs average IcBL',fontsize=20)

plt.subplot(122)
plotmont1('blicSpec_b_', 
         ['2', '8', '12', '18', '22', '28'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15', '20'], 
         [2, 6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(0, 31)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened blicSpec_b_ vs average IcBL',fontsize=20)
#pl.savefig(plotname)
```

Synthetic spectra have broad enough absorption features, but the Si II absorption around 6000 angstrom is too strong and the Ca II absorption around 8000 angstrom is not shown. 

At phase -5 and 0, synthetic spectra 'a' are more consistent with average spectra of IcBL. At phase 5 and 10, synthetic spectra 'b' are more consistent with average spectra of IcBL, especially at the expected positions of Si II 6355 and Ca II around 8000 angstrom. At phase 15, the synthetic spectrum 'a' nicely reproduces the 'w' feature of Fe II absorptions.

```python
def plotmont2(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        print spec+phase[i]+'.txt'
        w,f = readcoll.readcol(spec+phase[i]+'.txt',twod=False)
        f=f/np.mean(f)/5
        print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        dwbin=s.wlog[1:1024]-s.wlog[0:1023] # very important
        dwbin=np.append(dwbin[0], dwbin)/10
        pl.text(5200,shift[i]-0.5,phasemean[i]+'+/-2d',fontsize=20)
        
        if i == 0:
            pl.plot(w, f+shift[i], label=spec, color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f+shift[i], color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3)   
        pl.fill_between(s.wlog, (s.fmean-s.fsdev+1)*s.smean/dwbin+shift[i], (s.fmean+s.fsdev+1)*s.smean/dwbin
                        +shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont2('blicSpec_a_', 
         ['2', '6', '10', '16', '20', '26'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15', '20'], 
         [2, 6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(0, 31)
pl.legend(fontsize=20)
plt.tick_params(labelsize=20)
pl.title('Original blicSpec_a_ vs average IcBL',fontsize=20)

plt.subplot(122)
plotmont2('blicSpec_b_', 
         ['2', '8', '12', '18', '22', '28'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15', '20'], 
         [2, 6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(0, 31)
plt.tick_params(labelsize=20)
pl.title('Original blicSpec_b_ vs average IcBL',fontsize=20)
pl.legend(fontsize=20)
```

Above plots show average spectra with continuum. The synthetic spectra have bluer continuum than average IcBL. Since SLSNe have bluer continuum than IcBL, I compared synthetic spectra with average spectra of SLSNe below.

```python
def plotmont3(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        print spec+phase[i]+'.txt'
        w,f = readcoll.readcol(spec+phase[i]+'.txt',twod=False)
        f=f/np.mean(f[np.where((w > 3000) & (w < 8000))])
        print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        dwbin=s.wlog[1:1024]-s.wlog[0:1023] # very important
        dwbin=np.append(dwbin[0], dwbin)/8
        pl.text(5200,shift[i],phasemean[i]+'+/-2d',fontsize=20)

        if i == 0:
            pl.plot(w, f+shift[i], label=spec)
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3, label='average SLSNe')
        else:
            pl.plot(w, f+shift[i])
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3)   
        pl.fill_between(s.wlog, (s.fmean-s.fsdev+1)*s.smean/dwbin+shift[i], (s.fmean+s.fsdev+1)*s.smean/dwbin
                        +shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont3('blicSpec_a_', 
         ['6', '16',  '26'], 
         '../SLSNe/meanspec/meanspecSL_1specperSN_', 
         ['0', '10', '20'], 
         [2, 5, 8])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(2800, 7000)
plt.tick_params(labelsize=20)
pl.ylim(1, 12)
pl.title('Original blicSpec_a_ vs average SLSN',fontsize=20)
pl.legend(fontsize=20)

plt.subplot(122)
plotmont3('blicSpec_b_', 
         ['8', '18', '28'], 
         '../SLSNe/meanspec/meanspecSL_1specperSN_', 
         ['0', '10', '20'], 
         [2, 5, 8])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(2800, 7000)
pl.ylim(1, 12)
plt.tick_params(labelsize=20)
pl.title('Original blicSpec_b_ vs average SLSN',fontsize=20)
pl.legend(fontsize=20)
```

At phase 0, the synthetic spectra match the average spectra of SLSNe at most wavelenghes. At phase 10 and 20, the slope of synthetic spectra match the slope of the average spectra for wavelength larger than 4000 angstrom.
