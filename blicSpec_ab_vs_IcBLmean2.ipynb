This notebook
- compare synthetic spectra to average SN Ic-bl spectra without and with continuum
- show good matches between synthetic spectra and individual SN Ic-bl spectra from SNID
- compare sythetic LCs to those of SN Ic-bl shown in figure 8 of [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf) 

Main conclusion
- Synthetic spectra have broad features and continuum that are very similar to SN Ic-bl
- To be specific 
    - Model a: feature around 7100 angstrom at phases -5 and 0 is too strong; Ca II feature around 8000 is not shown at phases 5, 10 and 15; Fe II feature around 5000 angstrom is too strong at phase 15.
    - Model b: Fe II feature around 5000 angstrom is too strong at almost all phases.
- Synthetic LCs have a plateau phase, which is not shown in SN Ic-bl, as seen in figure 8 of [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf) 

```python
import numpy as np
from scipy.io.idl import readsav
import pylab as pl
import pandas as pd
import readcoll 
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from IPython.display import Image
from IPython.display import display
%matplotlib inline
```

# I. Plots below compare synthetic spectra to average SN Ic-bl spectra without and with continuum

## 1) Compare spectra without continuum

```python
def plotmont1(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        print spec+phase[i]+'_flat.txt'
        w,f = readcoll.readcol(spec+phase[i]+'_flat.txt',twod=False)
        print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        pl.text(4000,shift[i]+1,phasemean[i]+'+/-2d',fontsize=20)
        if i == 0:
            pl.plot(w, f*5+shift[i], label=spec, color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f*5+shift[i], color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3)    
        pl.fill_between(s.wlog, s.fmean*5-s.fsdev*5+shift[i], s.fmean*5+s.fsdev*5+shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont1('Sep2016/blic_mod02a_', 
         ['10', '16', '20', '26', '30'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05','00', '05', '10', '15'], 
         [6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 30)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened blic_a_ vs average IcBL',fontsize=20)

plt.subplot(122)
plotmont1('Sep2016/blic_mod02b_', 
         ['12', '18', '22', '28', '32'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15'], 
         [6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 30)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened blic_b_ vs average IcBL',fontsize=20)
#pl.savefig(plotname)
```

Model a: feature around 7100 angstrom at phases -5 and 0 is too strong; Ca II feature around 8000 is not shown at phases 5, 10 and 15; Fe II feature around 5000 angstrom is too strong at phase 15.

Model b: Fe II feature around 5000 angstrom is too strong at almost all phases.

## 2) Compare spectra with continuum

```python
def plotmont2(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        print spec+phase[i]+'.txt'
        w,f = readcoll.readcol(spec+phase[i]+'.txt',twod=False)
        f=f/np.mean(f)/4
        print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        dwbin=s.wlog[1:1024]-s.wlog[0:1023] # very important
        dwbin=np.append(dwbin[0], dwbin)/10
        pl.text(5200,shift[i]-0.5,phasemean[i]+'+/-2d',fontsize=20)
        
        if i == 0:
            pl.plot(w, f+shift[i], label=spec, color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f+shift[i], color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3)   
        pl.fill_between(s.wlog, (s.fmean-s.fsdev+1)*s.smean/dwbin+shift[i], (s.fmean+s.fsdev+1)*s.smean/dwbin
                        +shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plotmont2('Sep2016/blic_mod02a_', 
         ['10', '16', '20', '26', '30'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05','00', '05', '10', '15'], 
         [6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(0, 31)
pl.legend(fontsize=20)
plt.tick_params(labelsize=20)
pl.title('Original blicSpec_a_ vs average IcBL',fontsize=20)

plt.subplot(122)
plotmont2('Sep2016/blic_mod02b_', 
         ['12', '18', '22', '28', '32'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_1specperSN_', 
         ['-05', '00', '05', '10', '15'], 
         [6, 10, 14, 20, 24, 30])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(0, 31)
plt.tick_params(labelsize=20)
pl.title('Original blicSpec_b_ vs average IcBL',fontsize=20)
pl.legend(fontsize=20)
```

Above plots show average spectra with continuum. Continua are similar in synthetic spectra and average spectra.
